```python
from bs4 import BeautifulSoup
import requests

url = BeautifulSoup('https://www.doed.gov.np/license/54','html.parser')
response = requests.get(url)
data = response.text
soup = BeautifulSoup(data,'lxml')
tags = soup.table
tag = tags.find_all('tr')
a = []
for i in tag:
    s=i.get_text()
    a.append(s)
final = a[1:]
final_data = []
for i in final:
    f= i.split('\n')[1:16]
    final_data.append(f)
header = [['Sn','Project','Capacity(MW)','River','LicNo','Issue date','validity','promoter','Address','latitude1','latitude2','longitude1','longitude2','VDC/District','COD']]    
output_data = header + final_data
```

```python
import csv

f = open('hydropower.csv','w')
write = csv.writer(f)
for i in output_data:
    write.writerow(i)
f.close()
```
